```python
# Dependencies and Setup
import pandas as pd
```

```python
#load data
csv_load = ("../ResourcesFull/stats2021.csv")
db_2021 = pd.read_csv(csv_load, encoding = "ISO-8859-1")
```

```python
db_2021
```

```python
db_2021[["Player","Player_ID"]] = db_2021['Player'].str.split("\\", 1, expand=True)
db_2021.head()
```

```python
#removing players with less the half a season of games 
#(filtering players with too few games, players that were traded midseason
game_cutoff = 43
db_2021[db_2021["G"] > game_cutoff]
```

```python
db_2021_filtered = db_2021[db_2021["G"] > game_cutoff]
db_2021_filtered
```

```python
#player_names checks (now less vital with ids)
player_names = db_2021_filtered["Player"]
```

```python
#checking allstar player
player_names[player_names.str.match(".*\*")]
```

```python
#checking for special characters the may need hands on checks
player_names[player_names.str.match("[^A-Za-z0-9_]")]
```

```python
#correcting names
corrected_names = []
for player in player_names:
    if player.find("*") != (-1):
        player = player.replace("*","")
        print(player)
    elif player == "Ömer A??k":
        player = player.replace("A??k","Asik")
        print(player)
    corrected_names.append(player)
```
