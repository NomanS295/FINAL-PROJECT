```python
eight_df = pd.read_csv("stats2008_clean.csv", encoding = "ISO-8859-1")
nine_df = pd.read_csv("stats2009_clean.csv", encoding = "ISO-8859-1")
ten_df = pd.read_csv("stats2010_clean.csv", encoding = "ISO-8859-1")
eleven_df = pd.read_csv("stats2011_clean.csv", encoding = "ISO-8859-1")
twelve_df = pd.read_csv("stats2012_clean.csv", encoding = "ISO-8859-1")
thirteen_df = pd.read_csv("stats2013_clean.csv", encoding = "ISO-8859-1")
fourteen_df = pd.read_csv("stats2014_clean.csv", encoding = "ISO-8859-1")
fifteen_df = pd.read_csv("stats2015_clean.csv", encoding = "ISO-8859-1")
sixteen_df = pd.read_csv("stats2016_clean.csv", encoding = "ISO-8859-1")
seventeen_df = pd.read_csv("stats2017_clean.csv", encoding = "ISO-8859-1")
eighteen_df = pd.read_csv("stats2018_clean.csv", encoding = "ISO-8859-1")
nineteen_df = pd.read_csv("stats2019_clean.csv", encoding = "ISO-8859-1")
twenty_df = pd.read_csv("stats2020_clean.csv", encoding = "ISO-8859-1")
twentyone_df = pd.read_csv("stats2021_clean.csv", encoding = "ISO-8859-1")
```

```python
all_years = [eight_df, nine_df, ten_df, eleven_df, twelve_df, thirteen_df, fourteen_df, fifteen_df, sixteen_df, seventeen_df,
            eighteen_df, nineteen_df, twenty_df, twentyone_df]

combined_df = pd.concat(all_years)

combined_df
```

```python
combined_df = combined_df.sort_values(by=['Player_ID'])

combined_df
```

```python
combined_df.drop(['Player_ID', 'Player', 'Pos', 'Tm'], axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = combined_df.drop(['GMSCR'], axis=1)
```

```python
y = combined_df['GMSCR']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor()
```

```python
model.fit(X_train, y_train)
```

```python
preds = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, preds)
```
