# Machine Learning Model

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sbn
```

```python
data = "Resources/breast-cancer.csv"

cancer_df = pd.read_csv(data)

cancer_df
```

```python
cancer_df.describe()
```

```python
cancer_df.info()
```

```python
cancer_df['diagnosis'] = cancer_df['diagnosis'].map({'M': 1, 'B': 0})

cancer_df
```

```python
countBen = len(cancer_df[cancer_df.diagnosis == 0])
countMal = len(cancer_df[cancer_df.diagnosis == 1])
print("Percentage of Tumors that are Malignant: {:.2f}%".format((countMal / (len(cancer_df.diagnosis))*100)))
print("Percentage of Tumors that are Benign: {:.2f}%".format((countBen / (len(cancer_df.diagnosis))*100)))
```

## Logistic Regression

```python
cancer_df = cancer_df.drop(columns="id")

y = cancer_df["diagnosis"]
X = cancer_df.drop(columns="diagnosis")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, test_size = 0.20)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs', max_iter = 5690, random_state=1)
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, classification_report
matrix = confusion_matrix(y_test, y_pred)
print(matrix)
```

```python
report = classification_report(y_test, y_pred)
print(report)
```

```python
# Confusion Matrix
conf = confusion_matrix(y_test, y_pred)
df_conf = pd.DataFrame(conf, index = ['B', 'M'], columns = ['B', 'M'])
plt.figure(figsize = (6, 6))
sbn.heatmap(df_conf, annot=True, cmap='viridis')
```

## AdaBoost

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
from sklearn.ensemble import GradientBoostingClassifier
learning_rates = [0.05, 0.1, 0.25, 0.5, 0.75, 1]
for learning_rate in learning_rates:
    classifier = GradientBoostingClassifier(n_estimators=20,
    learning_rate=learning_rate,
    max_features=5,
    max_depth=3,
    random_state=0)
    classifier.fit(X_train, y_train)
    print("Learning rate: ", learning_rate)
    print("Accuracy score (training): {0:.3f}".format(
        classifier.score(
            X_train,
            y_train)))
    print("Accuracy score (validation): {0:.3f}".format(
           classifier.score(
            X_test,
            y_test)))
```

```python
#choosing learning rate of 0.25 for minimum risk of overfitting but shows different restults

classifier = GradientBoostingClassifier(n_estimators=20,
   learning_rate=0.25, max_features=5, max_depth=3, random_state=0)

classifier.fit(X_train, y_train)
predictions = classifier.predict(X_test)
```
